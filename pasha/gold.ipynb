## Данные
### info
+ <code>catalog_num</code> — каталожный номер
+ <code>letters</code> — буквы на монете
+ <code>mint</code> — монетный двор
+ <code>circulation</code> — тираж
+ <code>metal</code> — металл
+ <code>weight</code> — вес монеты (г.)
+ <code>pure_metal_content</code> — содержание чистого металла в монете (г.)
+ <code>diameter</code> — диаметр (мм.)
+ <code>obverse</code> — надпись на лицевой стороне
+ <code>reverse</code> — надпись на обратной стороне
+ <code>coin_name</code> — название монеты

### Sales
+ <code>state</code> — сохранность
 + G — Good
 + VG — Very Good
 + FALSE — Fine (должно быть F, но R посчитал это FALSE)
 + VF — Very Fine
 + XF — Extra Fine
 + AU — Almost Uncirculated
 + UNC — Uncirculated
 + Proof — Идеальное
+ <code>auction</code> — название аукциона
+ <code>buyer</code> — ник покупателя
+ <code>n_bid</code> — количество заявок на аукционе
+ <code>price</code> — цена сделки
+ <code>date</code> — дата сделки
+ <code>coin_name</code> — название монеты

```python
#Видно, что средняя цена действительно отличается (иногда сильно) в зависимости от аукциона для каждой монеты 
coin_names = set(sales["coin_name"])
auction_names = set(sales["auction"])
for coin_name in coin_names:
    print(f"Для монеты {coin_name}")
    for auction_name in auction_names:
        mean_price = round(sales[(sales["coin_name"] == coin_name) & (sales["auction"] == auction_name)]["price"].mean(), 3)
        print(f"Средняя цена продажи на аукционе {auction_name} равна: {mean_price}")
    print()
```

```python
def lr(name):
    df = sales[sales["coin_name"] == name]
    #features = df.drop(["date", "coin_name", "price"], axis=1).columns
    #df_scaled = sc.fit_transform(df.drop(["date", "coin_name", "price"], axis=1))
    #df_scaled = pd.DataFrame(df_scaled, columns=features)
    #df_scaled["price"] = list(df["price"])
    #display(df_scaled)
    #timestamp = df["date"]
    X = df.drop(["timestamp", "coin_name", "price"], axis=1)
    y = df["price"]
    #sc = StandardScaler()
    #X_scaled = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
    X_train, y_train = X[:int(X.shape[0] * 0.8)], y[:int(X.shape[0] * 0.8)]
    X_valid, y_valid = X[int(X.shape[0] * 0.8):], y[int(X.shape[0] * 0.8):]
    #lr
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    lr_rmse = round(mean_squared_error(y_valid, lr.predict(X_valid), squared=False), 3)
    plt.figure(figsize=(15, 7))
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=y_valid)
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=lr.predict(X_valid))
    plt.title(f"График предсказания стоимости монеты: {name}")
    plt.ylabel("Стоимость")
    plt.legend(["Actual", "LR predicted"])
    #оценка метрики
    print(f"Для криптовалюты {name} RMSE на LR: {lr_rmse}")
    print()
```

```python
lr("5 рублей, 1898 год")
```

```python
def rf(name):
    df = sales[sales["coin_name"] == name]
    #features = df.drop(["date", "coin_name", "price"], axis=1).columns
    #df_scaled = sc.fit_transform(df.drop(["date", "coin_name", "price"], axis=1))
    #df_scaled = pd.DataFrame(df_scaled, columns=features)
    #df_scaled["price"] = list(df["price"])
    #display(df_scaled)
    #timestamp = df["date"]
    X = df.drop(["timestamp", "coin_name", "price"], axis=1)
    y = df["price"]
    #sc = StandardScaler()
    #X_scaled = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
    X_train, y_train = X[:int(X.shape[0] * 0.8)], y[:int(X.shape[0] * 0.8)]
    X_valid, y_valid = X[int(X.shape[0] * 0.8):], y[int(X.shape[0] * 0.8):]
    #rf
    rf = RandomForestRegressor()
    params = {
        "n_estimators": range(50, 301, 50),
        "max_depth": range(3, 23, 2),
        "max_features": ["sqrt", "log2"]
    }
    grid = GridSearchCV(rf, param_grid=params, n_jobs=-1)
    grid.fit(X_train, y_train)
    grid_best = grid.best_estimator_
    print(f"Лучшие параметры на кросс-валидации для монеты '{name}': {grid.best_params_}")
    rf_rmse = round(mean_squared_error(y_valid, grid_best.predict(X_valid), squared=False), 3)
    plt.figure(figsize=(15, 7))
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=y_valid)
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=grid_best.predict(X_valid))
    plt.title(f"График предсказания стоимости монеты: {name}")
    plt.ylabel("Стоимость")
    plt.legend(["Actual", "RF predicted"])
    #оценка метрики
    print(f"Для криптовалюты {name} RMSE на RF: {rf_rmse}")
    print()
```

```python
%%time
rf("5 рублей, 1898 год")
```

```python
def polynom(name):
    df = sales[sales["coin_name"] == name]
    #features = df.drop(["date", "coin_name", "price"], axis=1).columns
    #df_scaled = sc.fit_transform(df.drop(["date", "coin_name", "price"], axis=1))
    #df_scaled = pd.DataFrame(df_scaled, columns=features)
    #df_scaled["price"] = list(df["price"])
    #display(df_scaled)
    #timestamp = df["date"]
    X = df.drop(["timestamp", "coin_name", "price"], axis=1)
    y = df["price"]
    #sc = StandardScaler()
    #X_scaled = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
    X_train, y_train = X[:int(X.shape[0] * 0.8)], y[:int(X.shape[0] * 0.8)]
    X_valid, y_valid = X[int(X.shape[0] * 0.8):], y[int(X.shape[0] * 0.8):]
    #polynom
    pol = PolynomialFeatures(degree=2)
    X_train_pol = pol.fit_transform(X_train)
    X_valid_pol = pol.transform(X_valid)
    lr_pol = LinearRegression()
    lr_pol.fit(X_train_pol, y_train)
    pol_rmse = round(mean_squared_error(y_valid, lr_pol.predict(X_valid_pol), squared=False), 3)
    plt.figure(figsize=(15, 7))
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=y_valid)
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=lr_pol.predict(X_valid_pol))
    plt.title(f"График предсказания стоимости монеты: {name}")
    plt.ylabel("Стоимость")
    plt.legend(["Actual", "POLYNOM predicted"])
    #оценка метрики
    print(f"Для криптовалюты {name} RMSE на Polynom: {pol_rmse}")
    print()
```

```python
def lr_all(df):
    X = df.drop(["coin_name", "timestamp", "price", "dollar_value", "price_usd", "n_bid", "metal_content", "diameter"], axis=1)
    y = df["price_usd"]
    features = ["n_bid", "metal_content", "diameter"]
    df_for_scale = df[["n_bid", "metal_content", "diameter"]]
    sc = StandardScaler()
    df_scaled = sc.fit_transform(df_for_scale)
    df_scaled = pd.DataFrame(df_scaled, columns=features)
    X = X.join(df_scaled)
    X_train, y_train = X[:int(X.shape[0] * 0.8)], y[:int(X.shape[0] * 0.8)]
    X_valid, y_valid = X[int(X.shape[0] * 0.8):], y[int(X.shape[0] * 0.8):]
    #lr
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    lr_rmse = round(mean_squared_error(y_valid, lr.predict(X_valid), squared=False), 3)
    plt.figure(figsize=(15, 7))
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=y_valid)
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=lr.predict(X_valid))
    plt.ylabel("Стоимость")
    plt.legend(["Actual", "LR predicted"])
    #оценка метрики
    print()
    print(f"Среднеквадратическое отклонение RMSE на LR: {lr_rmse}")
```

```python
def rf_all(df):
    X = df.drop(["coin_name", "timestamp", "price", "dollar_value", "price_usd"], axis=1)
    y = df["price_usd"]
    X_train, y_train = X[:int(X.shape[0] * 0.8)], y[:int(X.shape[0] * 0.8)]
    X_valid, y_valid = X[int(X.shape[0] * 0.8):], y[int(X.shape[0] * 0.8):]
    #rf
    rf = RandomForestRegressor()
    params = {
        "n_estimators": range(50, 301, 50),
        "max_depth": range(3, 23, 2),
        "max_features": ["sqrt", "log2"]
    }
    grid = GridSearchCV(rf, param_grid=params, n_jobs=-1)
    grid.fit(X_train, y_train)
    grid_best = grid.best_estimator_
    print(f"Лучшие параметры на кросс-валидации для монеты: {grid.best_params_}")
    rf_rmse = round(mean_squared_error(y_valid, grid_best.predict(X_valid), squared=False), 3)
    plt.figure(figsize=(15, 7))
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=y_valid)
    sns.lineplot(x=df["timestamp"][int(X.shape[0] * 0.8):], y=grid_best.predict(X_valid))
    plt.ylabel("Стоимость")
    plt.legend(["Actual", "RF predicted"])
    #оценка метрики
    print(f"Среднеквадратическое отклонение RMSE на LR: {rf_rmse}")
```
